```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
df=pd.read_csv('iris.csv',index_col='Id')
df.head()
```

```python
sns.distplot(df['Petal Length (cm)'],kde=False)
```

```python
# KDE plot 
sns.kdeplot(data=df['Petal Length (cm)'], shade=True)
```

```python
# KDE plot 
sns.kdeplot(data=df['Sepal Length (cm)'], shade=True)
```

```python
sns.jointplot(x=df['Petal Length (cm)'], y=df['Sepal Width (cm)'], kind="kde")
```

```python
df_set=pd.read_csv('iris_setosa.csv',index_col='Id')
df_ver=pd.read_csv('iris_versicolor.csv',index_col='Id')
df_vir=pd.read_csv('iris_virginica.csv',index_col='Id')

df_vir.head()
```

```python
sns.distplot(df_vir['Petal Length (cm)'],kde=False)
```

```python
sns.distplot(a=df_set['Petal Length (cm)'], label="Iris-setosa", kde=False)
sns.distplot(a=df_ver['Petal Length (cm)'], label="Iris-versicolor", kde=False)
sns.distplot(a=df_vir['Petal Length (cm)'], label="Iris-virginica", kde=False)

# Add title
plt.title("Histogram of Petal Lengths, by Species")

# Force legend to appear
plt.legend()
```

```python
sns.kdeplot(data=df_set['Petal Length (cm)'], label="Iris-setosa", shade=True)
sns.kdeplot(data=df_ver['Petal Length (cm)'], label="Iris-versicolor",shade=True)
sns.kdeplot(data=df_vir['Petal Length (cm)'], label="Iris-virginica", shade=True)

# Add title
plt.title("Histogram of Petal Lengths, by Species")

# Force legend to appear
plt.legend()
```
