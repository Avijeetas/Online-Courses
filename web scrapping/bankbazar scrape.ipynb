```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
reviews=[]
user=[]
banks=[]
def build():
    for i in range(1,51):
        url='https://www.bankbazaar.com/reviews.html?reviewPageNumber={}'.format(i)
        page=requests.get(url)    
        soup= BeautifulSoup(page.text,'html.parser')
        divv=soup.find_all('li',{'class':'review-box'})
        
        for i in divv:
            x=i.find(class_='text_here review-desc-more').text.strip()
            reviews.append(x)
            x=i.find(class_='js-author-name').text.strip()
            user.append(x)
            x=i.find('img').get('alt')
            banks.append(x)    
```

```python
build()
```

```python
all_data=[]

for text,usr,bank in zip(reviews,user,banks):
    all_data.append({'Review_text':text,'User':usr,'Bank':bank})
df=pd.DataFrame(all_data)
df
```
