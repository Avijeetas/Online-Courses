```python
df = pd.read_excel('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DV0101EN/labs/Data_Files/Canada.xlsx',
                       sheet_name='Canada by Citizenship',
                       skiprows=range(20),
                       skipfooter=2
                      )

print('Data downloaded and read into a dataframe!')
```

```python
df.head()
```

```python
nw_df=df.drop(['Coverage','Type','AREA','REG','DEV'],axis=1)
```

```python
nw_df['Total']=nw_df.sum(axis=1)
```

```python
nw_df.rename(columns={'OdName':'Country', 'AreaName':'Continent','RegName':'Region'}, inplace=True)

# let's view the first five elements and see how the dataframe was changed
nw_df.head()
```

```python
nw_df.set_index('Country', inplace=True)
nw_df.index
```

```python
years = list(map(str, range(1980, 2014)))
nw_df.head() # returns a series
```

```python
### type your answer here
'''

 # Step 1: Get the dataset. Recall that we created a Total column that calculates the cumulative immigration by country. \\ We will sort on this column to get our top 5 countries using pandas sort_values() method.
# inplace = True paramemter saves the changes to the original df_can dataframe
nw_df.sort_values(by='Total', ascending=False, axis=0, inplace=True)
# get the top 5 entries
df_top5 = nw_df.head(5)
df_top5
'''
nw_df.index.name = None
#nw_df1=nw_df[['Country', 1980, 1981, 1982, 1983, 1984, 1985]]
nw_df
```

```python
nw_df.columns = list(map(str, nw_df.columns))
nw_df
```

```python
nw_df1=nw_df
nw_df1.sort_values(by='Total', ascending=False, axis=0, inplace=True)
top5=nw_df1.head(5)
top5=top5[years].transpose()
top5.index=top5.index.map(int)

top5.plot(kind='line',figsize=(14,8))
plt.xlabel('Years')
plt.ylabel('Numbers of Immigrants')
plt.title('Immigrants of top 5 countries')

plt.show()
```

### AREA PLOTS


```python
top5.plot(kind='area', 
             stacked=True,
             figsize=(20, 10), # pass a tuple (x, y) size
             )

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

```python
# generate histogram
nw_df.loc[['Denmark', 'Norway', 'Sweden'], years].plot.hist()
```

```python
# transpose dataframe
df_t = nw_df.loc[['Denmark', 'Norway', 'Sweden'], years].transpose()
df_t.head()
```

```python
# generate histogram
df_t.plot(kind='hist', figsize=(10, 6))

plt.title('Histogram of Immigration from Denmark, Norway, and Sweden from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')

plt.show()
```

```python
# let's get the x-tick values
count, bin_edges = np.histogram(df_t, 15)

# un-stacked histogram
df_t.plot(kind ='hist', 
          figsize=(10, 6),
          bins=15,
          alpha=0.6,
          xticks=bin_edges,
          color=['coral', 'darkslateblue', 'mediumseagreen']
         )

plt.title('Histogram of Immigration from Denmark, Norway, and Sweden from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')

plt.show()
```

```python
count, bin_edges = np.histogram(df_t, 15)
xmin = bin_edges[0] - 10   #  first bin value is 31.0, adding buffer of 10 for aesthetic purposes 
xmax = bin_edges[-1] + 10  #  last bin value is 308.0, adding buffer of 10 for aesthetic purposes

# stacked Histogram
df_t.plot(kind='hist',
          figsize=(10, 6), 
          bins=15,
          xticks=bin_edges,
          color=['coral', 'darkslateblue', 'mediumseagreen'],
          stacked=True,
          xlim=(xmin, xmax)
         )

plt.title('Histogram of Immigration from Denmark, Norway, and Sweden from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants') 

plt.show()
```

```python
nw_df.sort_values(by='Total', ascending=True, inplace=True)
df_top15 = nw_df['Total'].tail(15)
df_top15.plot(kind='barh', figsize=(12, 12), color='steelblue')
plt.xlabel('Number of Immigrants')
plt.title('Top 15 Conuntries Contributing to the Immigration to Canada between 1980 - 2013')
for index, value in enumerate(df_top15): 
    label = format(int(value), ',') # format int with commas
    
    plt.annotate(label, xy=(value - 47000, index - 0.10), color='black')
plt.show()
```
