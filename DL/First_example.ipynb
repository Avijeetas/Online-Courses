```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
obs=1000

xs=np.random.uniform(low=-10, high=10,size=(obs,1))
zs=np.random.uniform(low=-10, high=10,size=(obs,1))

inputs=np.column_stack((xs,zs))
print(inputs.shape)
```

```python
#Create  the targets we'll aim at

noise=np.random.uniform(-1,1,(obs,1))
targets= 2*xs -3*zs  + 5 + noise

print(targets.shape)
```

```python
#plot the training data

targets = targets.reshape(obs,)
flg=plt.figure()
ax=flg.add_subplot(111,projection='3d')

ax.plot(xs,zs,targets)
ax.view_init(azim=100)
plt.show()
targets = targets.reshape(obs,1)
```
